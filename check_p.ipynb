```python
import random

num = random.randint(1, 101)
count = 0
while True:
    guess = int(input("Guess the number generated: "))
    if num == guess:
        count += 1
        print("Congrats, you are smart!")
        break 
    elif num < guess :
        count += 1
        print("wrong , number too high try again .")
    else : 
        count += 1
        print("wrong , number too low try again ")

print(f"you got it on try : {count}")

```
